```python
from enum import Enum

class CommonType(Enum):
    
    @classmethod
    def values(cls) -> list:
        return [item.value for item in cls]


class WorkclassType(CommonType):
    PRIVATE = 'Private'
    SELF_EMP_NOT_INC = 'Self-emp-not-inc'
    SELF_EMP_INC = 'Self-emp-inc'
    FEDERAL_GOV = 'Federal-gov'
    LOVAL_GOV = 'Local-gov'
    STATE_GOV = 'State-gov'
    WITHOUT_PAY = 'Without-pay'
    NEVER_WORKED = 'Never-worked'


class EducationType(CommonType):
    BACHELORS = 'Bachelors'
    SOME_COLLEGE = 'Some-college'
    _11TH = '11th'
    HD_GRAD = 'HS-grad'
    PROF_SCHOOL = 'Prof-school'
    ASSOC_ACDM = 'Assoc-acdm'
    ASSOC_VOC = 'Assoc-voc'
    _9TH = '9th'
    _7TH_8TH = '7th-8th'
    _12TH = '12th'
    MASTERS = 'Masters'
    _1ST_4TH = '1st-4th'
    _10TH = '10th'
    DOCTORATE = 'Doctorate'
    _5TH_6TH = '5th-6th'
    PRESCHOOL = 'Preschool'


class MaritalstatusType(CommonType):
    MARRIED_CIV_SPOUSE = 'Married-civ-spouse'
    DIVORCED = 'Divorced'
    NEVER_MARRIED = 'Never-married'
    SEPARATED = 'Separated'
    WIDOWED = 'Widowed'
    MARRIED_SPOUSE_ABSENT = 'Married-spouse-absent'
    MARRIED_AF_SPOUSE = 'Married-AF-spouse'


class OccupationType(CommonType):
    TECH_SUPPORT = 'Tech-support'
    CRAFT_REPAIR = 'Craft-repair'
    OTHER_SERVICE = 'Other-service'
    SALES = 'Sales'
    EXEC_MANAGERIAL = 'Exec-managerial'
    PROF_SPECIALTY = 'Prof-specialty'
    HANDLERS_CLEANERS = 'Handlers-cleaners'
    MACHINE_OP_INSPCT = 'Machine-op-inspct'
    ADM_CLERICAL = 'Adm-clerical'
    FARMING_FISHING = 'Farming-fishing'
    TRANSPORT_MOVING = 'Transport-moving'
    PRIV_HOUSE_SERV = 'Priv-house-serv'
    PROTECTIVE_SERV = 'Protective-serv'
    ARMED_FORCES = 'Armed-Forces'


class RelationshipType(CommonType):
    WIFE = 'Wife'
    OWN_CHILD = 'Own-child'
    HUSBAND = 'Husband'
    NOT_IN_FAMILY = 'Not-in-family'
    OTHER_RELATIVE = 'Other-relative'
    UNMARRIED = 'Unmarried'


class RaceType(CommonType):
    WHITE = 'White'
    ASIAN_PAC_ISLANDER = 'Asian-Pac-Islander'
    AMER_INDIAN_ESKIMO = 'Amer-Indian-Eskimo'
    OTHER = 'Other'
    BLACK = 'Black'


class SexType(CommonType):
    FEMALE = 'Female'
    MALE = 'Male'


class NativecountryType(CommonType):
    UNITED_STATES = 'United-States'
    CAMBODIA = 'Cambodia'
    ENGLAND = 'England'
    PUERTO_RICO = 'Puerto-Rico'
    CANADA = 'Canada'
    GERMANY = 'Germany'
    OUTLYING_US = 'Outlying-US(Guam-USVI-etc)'
    INDIA = 'India'
    JAPAN = 'Japan'
    GREECE = 'Greece'
    SOUTH = 'South'
    CHINA = 'China'
    CUBA = 'Cuba'
    IRAN = 'Iran'
    HONDURAS = 'Honduras'
    PHILIPPINES = 'Philippines'
    ITALY = 'Italy'
    POLAND = 'Poland'
    JAMAICA = 'Jamaica'
    VIETNAM = 'Vietnam'
    MEXICO = 'Mexico'
    PORTUGAL = 'Portugal'
    IRELAND = 'Ireland'
    FRANCE = 'France'
    DOMINICAN_REPUBLIC = 'Dominican-Republic'
    LAOS = 'Laos'
    ECUADOR = 'Ecuador'
    TAIWAN = 'Taiwan'
    HAITI = 'Haiti'
    COLUMBIA = 'Columbia'
    HUNGARY = 'Hungary'
    GUATEMALA = 'Guatemala'
    NICARAGUA = 'Nicaragua'
    SCOTLAND = 'Scotland'
    THAILAND = 'Thailand'
    YUGOSLAVIA = 'Yugoslavia'
    EL_SALVADOR = 'El-Salvador'
    TRINADAD_TOBAGO = 'Trinadad&Tobago'
    PERU = 'Peru'
    HONG = 'Hong'
    HOLAND_NETHERLANDS = 'Holand-Netherlands'


class ClassType(CommonType):
    GT_50K = '>50K'
    LTE_50K = '<=50K'
```

```python
import pandas as pd
columns = [
        'age',
        'workclass',
        'fnlwgt',
        'education',
        'education-num',
        'marital-status',
        'occupation',
        'relationship',
        'race',
        'sex',
        'capital-gain',
        'capital-loss',
        'hours-per-week',
        'native-country',
        'class',
    ]

data = pd.read_csv(
    filepath_or_buffer='Adult.data',
    header=None,
    names=columns
)
data_test = pd.read_csv(
    filepath_or_buffer='Adult.test',
    header=None,
    names=columns,
    skiprows=1
)
df = pd.concat([data, data_test]).reset_index(drop=True)
```

```python
# Verificando os tipos de dados das colunas
df.info()
```

```python
# Como os tipos deveriam ser

"""
              :    IDEAL   :  REAL  :       
age           : continuous : string : 
workclass     : string     : string : OK
fnlwgt        : continuous : string : 
education     : string     : string : OK
education-num : continuous : int64  : OK
marital-status: string     : string : OK
occupation    : string     : string : OK
relationship  : string     : string : OK
race          : string     : string : OK
sex           : string     : string : OK
capital-gain  : continuous : string :
capital-loss  : continuous : int64  : OK
hours-per-week: continuous : string
native-country: string     : string : OK
class         : string     : string : OK
"""
```

```python
# Analisar porque as colunas que deveriam ser inteiro mas estão como string
# São as colunas 'age', 'fnlwgt', 'capital-gain', 'hours-per-week'

sample = []
for column in ['age', 'fnlwgt', 'capital-gain', 'hours-per-week']:
    for row in df[column]:
        try:
            int(row)
        except ValueError:
            sample.append(
                {
                    "index": int(df[df[column] == row].index.values),
                    "column": column,
                    "value": row
                }
            )
sample
```

```python
# No resultado acima vemos que alguns valores que deveriam ser int mas, são letras.
# Como são pouquissimos valores na massa de dados não vejo problema em descartar esse registros
for item in sample:
    df = df.drop(index=item['index'])
df
```

```python
# resetar o index
df = df.reset_index(drop=True)
df
```

```python
# mudar o tipo de dados de string para int das colunas analisadas
for column in ['age', 'fnlwgt', 'capital-gain', 'hours-per-week']:
   df[column] = df[column].astype('int')
df.info()
```

## O próximo passo é avaliar se valores das outras colunas que são do tipo string estão dentro dos valores estabelecidos

```python
# Primeiro vamos verificar se os valores batem coluna a coluna se não bater vou expor esse valor para sabermos o que é

# monta um array com tuplas que contém o nome da coluna do tipo string com o seu Enum, definido em contants
columns_string = [(column, f'{column.replace("-", "").capitalize()}Type') for column in df.columns if df[column].dtypes == 'object']

# Verifica qual o valor que está errado em cada coluna
invalid_values = []
for column in columns_string:
    values = df[column[0]].str.strip().unique()
    for value in values: 
        if value not in eval(column[1]).values():
            invalid_values.append(
                {
                    "column": column[0],
                    "value": value
                }
            )
invalid_values
```

```python
# No resultado acima vemos que tem o caracter '?' presente em algumas colunas
# E na coluna 'class' temos valores fora do padão
# Removendo os espaços vazios nas colunas do tipo string
for col in df.columns:
    if df[col].dtypes == 'object':
        df[col] = df[col].str.strip()
```

```python
df['class'].unique()
```

```python
# Vemos que na columa 'class' tem 2 valores que tem um ponto no final do valor.
# Vamos remover esse ponto para deixa no padrão esperado
df['class'] = df['class'].replace('<=50K.', '<=50K')
df['class'] = df['class'].replace('>50K.', '>50K')

df['class'].unique()
```

```python
# Todos os valores que estão com '?' nas colunas 'workclass', 'occupation', 'native-country' vou substituir por NaN
# Para uma analise posterior
import numpy as np
df.replace('?', np.NaN, inplace=True)
df.info()
```

```python
# O proximo passo é verificar o quanto os valores null representam em cada coluna
# A tomada de decisaão vai ser a seguinte. Caso a incidencia dos valores null
# seja < 5% vou deletar as linhas. Caso seja superior a 5% vou analisar para
# poder substituir pelo valor de maior incidencia, caso não seja possível vou exluir também

columns = ['workclass', 'occupation', 'native-country']

df_null_values = df.isnull()

out = []

for column in columns:
    series = df_null_values[column]\
                .value_counts(normalize=True)\
                .sort_index(ascending=False)
    out.append({
        "column": column,
        "null_values": f'{series.values[0] * 100 :.2f} %'
    })
pd.DataFrame(out)
```

```python
# Como terá remoção de registros da tabela vou fazer uma cópia para não alterar a frequencia dos valos

df_copy = df.copy()
```

```python
# como na coluna 'native-country' os valores nulos estão abaixa dos 5% vou excluir

ids = df_copy[df_copy['native-country'].isnull()].index.values
df_copy = df_copy.drop(index=list(ids))

df_copy
```

```python
# Como a coluna workclass tem > 5% dos valores nulos vamos verificar se conseguimos substituir pelo indice de maior frequencia

serie = df['workclass'].value_counts()


index = serie.index.to_list()
values = serie.to_list()
percentage = [f'{(value / sum(values) * 100) :.2f} %'  for value in values]

pd.DataFrame({'quantity': values, 'frequency (%)': percentage}, index=index)
```

```python
# Como a indidencia do valor 'Private' é muito alta não tem grandes problemas em substituirmos os valores nulos por 'Private'

df_copy['workclass'] = df_copy['workclass'].replace(np.NaN, 'Private')
```

```python
# Vou fazer a mesma análise para a coluna 'occupation'

serie = df['occupation'].value_counts()


index = serie.index.to_list()
values = serie.to_list()
percentage = [f'{(value / sum(values) * 100) :.2f} %'  for value in values]

pd.DataFrame({'quantity': values, 'frequency (%)': percentage}, index=index)
```

```python
# Como não temos um valor com frequencia preponderante os 3 primeiros tem frequencia
# bem próxima vou exluir as linhas já que os valores não são muito maiores do que 5%

ids = df_copy[df_copy['occupation'].isnull()].index.values
df_copy = df_copy.drop(index=list(ids))
```

```python
# Agora já que analisamos todas as colunas vamos remover as linhas duplicadas e resetar o index

df_copy = df_copy.drop_duplicates()

df_copy = df_copy.reset_index(drop=True)
```

```python
# Agora vou fazer uma analise em busca de outliers para as colunas de valores numéricos
# para excluir as linhas casos seja um outlier para não causar anomalias em análises futuras
# pela interatividade da ferramenta utilizarei a função express da biblioteca plotly
import plotly.express as px
df_copy.describe()
```

```python
# Irei verificar a distribuição dos dados se seguir uma distribuição normal utilizarei método Z-score para descobrir os valores outliers
# Caso não siga uma distribuição normal utilizarei o método de IQR para descobrir os outliers

fig = px.histogram(df_copy, x='age')

def find_outliers_IQR(df):
   q1=df.quantile(0.25)
   q3=df.quantile(0.75)
   IQR=q3-q1
   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
   return outliers

find_outliers_IQR(df_copy['fnlwgt'])
```
