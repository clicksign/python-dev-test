```python
import pandas as pd
```

Lista de atributos retirada do arquivo de descrição dos dados.

```python
cols = ["age", 
        "workclass",
        "fnlwgt",
        "education",
        "education-num",
        "marital-status",
        "occupation",
        "relationship",
        "race",
        "sex",
        "capital-gain",
        "capital-loss",
        "hours-per-week",
        "native-country",
        "class",
       ]
```

Substituindo o "-" por "_\" nos nomes dos atributos para salvar no banco de dados.

```python
cols = [attribute.replace("-", "_") for attribute in cols]
cols
```

Lendo o arquivo "Adult.data" e atribuindo seus respectivos atributos

```python
data = pd.read_csv("./data/Adult.data", delimiter=",", header=None, names=cols)
data
```

Lendo o arquivo "Adult.test" e atribuindo seus respectivos atributos.
Obs: Foi necessário pular a primeira linha que continha os caracteres "|1x3 Cross validator", acredito representar que essa base de dados foi retirada de 1/3 da base original para teste, outro fator que contribui com essa conclusão é a base de teste representar cerca de 1/3 do seu tamanho somado com o da base anterior. Representando o 1x3 retirado para teste através de validação cruzada.

```python
test = pd.read_csv("./data/Adult.test", delimiter=",", skiprows=1, header=None, names=cols)

test
```

Unindo os dados novamente para facilitar o estudo e a inserção no banco de dados.

```python
df = pd.concat([data, test], ignore_index=True, sort=False)
df.head()
```

Verificando informações úteis sobre os dados como valores nulos e o tipo.

```python
df.info()
```

Atribuindo o tipo numérico aos atributos seguindo a descrição fornecida.

```python
numeric_columns = ["age", "fnlwgt", "education_num", "capital_gain", "capital_loss", "hours_per_week"]

df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors="coerce")

df.head()
```

Verificando os atribulos nulos gerados pela transformação para numérico.

```python
df.isnull().sum()
```

Pensando como uma base de dados para aplicação de machine learning é recomendado substituir os valores faltantes pela média dos valores de mesma classe.

```python
for column in df.columns[df.isnull().any()]:
    df[column] = df[column].fillna(df.groupby("class")[column].transform("mean"))
```

Verificando se funcionou.

```python
df.isnull().sum()
```

Seguindo a propria descrição os atributos "marital_status" e "relationship" acabam sendo equivalentes, sendo redundante manter os dois na base de dados, podendo eliminar um deles.

```python
df = df.drop(["relationship"], axis=1)
```

O dataset agora já está em um formato adequado para salvarmos no banco de dados.
